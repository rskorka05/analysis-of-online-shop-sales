```python
df['Miesiac'] = df['Data_zakupu'].dt.to_period('M')
df
```

```python
miesiac_przychod = df.groupby('Miesiac')['Kwota_zamowienia'].sum()

miesiac_przychod
```

```python
plt.figure(figsize=(10,5))
plt.plot(miesiac_przychod.index.astype(str), miesiac_przychod.values, marker='o');
plt.title("Miesięczny przychód sklepu", fontsize=20)
plt.xlabel("Miesiąc", fontsize=15)
plt.ylabel("Przychód [mln zł]", fontsize=15)
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

Na wykresie widać, że miesięczny przychód sklepu utrzymuje się na stabilnym poziomie (ok. 1,6-1,9 mln zł). Najwyższe wartości pojawiają się w grudniu i styczniu, co sugeruje wzrost sprzedaży związany z okresem świątecznym i noworocznym. Z kolei w okresie wakacyjnym (lipiec-sierpień) przychody spadają, prawdopodobnie ze względu na mniejszą aktywność zakupową klientów.

```python
metody_platnosci = df[["Metoda_platnosci"]].value_counts()

metody_platnosci
```

```python
suma_transakcji = df[["Metoda_platnosci"]].value_counts().sum()

suma_transakcji
```

```python
procent_metody_platnosci = ((metody_platnosci / suma_transakcji) * 100).round(2)

procent_metody_platnosci
```

```python
plt.figure(figsize=(6,6))
plt.pie(
    metody_platnosci,
    autopct='%1.1f%%',
    startangle=90,
)

procent_metody_platnosci.index = procent_metody_platnosci.index.get_level_values(0)

plt.legend(
    procent_metody_platnosci.index,
    title="Metody płatności",
    bbox_to_anchor=(1, 0, 0, 1)
)

plt.title("Udział procentowy metod płatności", fontsize=15)
plt.show()
```

```python
suma_przychodow_woj = df[["Wojewodztwo", "Kwota_zamowienia"]].groupby("Wojewodztwo").sum().rename(
    columns={"Kwota_zamowienia": "suma_przychodow"}
).sort_values("suma_przychodow")

suma_przychodow_woj
```

```python
plt.figure(figsize=(15,5))
plt.bar(suma_przychodow_woj.index, suma_przychodow_woj["suma_przychodow"], color="#f5b042")
plt.title("Suma przychodow wg wojewodztwa", fontsize=20)
plt.xlabel("Wojewodztwo", fontsize=15)
plt.ylabel("Suma przychodow [mln zł]", fontsize=15)
plt.xticks(rotation=45)
plt.show()
```

```python
ilosc_zamowien_woj = df[["Wojewodztwo"]].groupby("Wojewodztwo").value_counts().sort_values()

ilosc_zamowien_woj
```

```python
plt.figure(figsize=(15,5))
plt.bar(ilosc_zamowien_woj.index, ilosc_zamowien_woj, color="#42f5d7")
plt.title("Ilość zamówień wg wojewodztwa", fontsize=20)
plt.xlabel("Wojewodztwo", fontsize=15)
plt.ylabel("Liczba zamówień", fontsize=15)
plt.xticks(rotation=45)
plt.show()
```

Najwięcej zamówień pochodzi z województwa Pomorskiego, a najmniej z województwa Opolskiego
